Dropping unnecessary columns.

```python
drop = ['photoUrl','playerUrl','Loan Date End']
df = df.drop(columns= drop)
```

Renaming some columns for consistency.

```python
df = df.rename(columns={'↓OVA':'OVA'})
```

Displaying the Types of Every Variable

As we can see, the variables (Height, Weight) are both object types despite being numerical numbers. We need to change the type of both of these variables into "int" type. We also need to convert other values in order to have uniform measurements.

```python
height = []

for i in df["Height"]:
    if i.endswith("cm") == True:
        i = i.replace('cm','')
        height.append(i)
    else:
        i = i.replace("\'", " ").replace("\"", "").split()
        a = int(i[0])*12 + int(i[1])
        cm = round(a*2.54)
        height.append(cm)
```

Renaming the height column with proper unit.

```python
df["Height"] = height
df.rename(columns={'Height': 'Height in cm'}, inplace=True)
```

Changing the data type of the height column.

```python
df = df.astype({'Height in cm':'int64'})
```

Changing weight in lbs into weight in kg.

```python
weight = []

for i in df["Weight"]:
    if i.endswith("lbs") == True:
        i = i.replace('lbs','')
        a = round(int(i)*.453592)
        weight.append(a)
    else:
        i = i.replace('kg','')
        weight.append(i)
```

Renaming weight column with proper unit.

```python
df["Weight"] = weight
df.rename(columns={'Weight':'Weight in kg'}, inplace=True)
```

Changing the data type of the weight column. 

```python
df = df.astype({'Weight in kg':'int64'})
```

Displaying the data types.

```python
df.info()
```

Both weight and height columns are now in their appropriate data types.

Next column to be cleaned is "Club". There are some unnecessary newline characters that we need to omit.

```python
df['Club'] = df['Club'].str.replace('\n\n\n\n','')
```

```python
df.head()
```

Splitting the Joined column into Year, Month and Day columns.

```python
Joined = []
for i in df['Joined']:
    i = i.replace("-"," ").split()
    Joined.append(i)
```

```python
Day = []
Month = []
Year = []
for i in Joined:
    a = i[0]
    b = i[1]
    c = i[2]
    Day.append(a)
    Month.append(b)
    Year.append(c)
```

```python
df['Day of Joining'] = Day
df['Month of Joining'] = Month
df['Year of Joining'] = Year
```

```python
df.head()
```

Dropping Joined column.

```python
df = df.drop(['Joined'], axis=1)
```

Changing the Value, Wage and Release columns into integers.

```python
Value = []
for i in df['Value']:
    if "M" in i:
        a = i.replace("€",'')
        b = a.replace("M",'')
        b = float(b) * 1000000
        Value.append(b)
    elif "K" in i:
        a = i.replace("€",'')
        b = a.replace("K",'')
        b = float(b) * 1000
        Value.append(b)
    else:
        i = i.replace("€",'')
        Value.append(i)
```

```python
df['Value'] = Value
df.rename(columns={'Value':'Value in €'}, inplace=True)
```

```python
Wage = []
for i in df['Wage']:
    if "K" in i:
        a = i.replace("€",'')
        b = a.replace("K",'')
        b = float(b) * 1000
        Wage.append(b)
    else:
        i = i.replace("€",'')
        Wage.append(i)

```

```python
df['Wage'] = Wage
df.rename(columns={'Wage':'Wage in €'}, inplace=True)
```

```python
Release = []
for i in df['Release Clause']:
    if "M" in i:
        a = i.replace("€",'')
        b = a.replace("M",'')
        b = float(b) * 1000000
        Release.append(b)
    elif "K" in i:
        a = i.replace("€",'')
        b = a.replace("K",'')
        b = float(b) * 1000
        Release.append(b)
    else:
        i = i.replace("€",'')
        Release.append(i)
```

```python
df['Release Clause'] = Release
df.rename(columns={'Release Clause':'Release Clause in €'}, inplace=True)
```

Some columns have Star ★ characters in their rows. We also need to omit these characters.

```python
def star(x):
    x = x.replace('★','')
    return x
```

```python
df['W/F'] = df['W/F'].apply(star)
df['SM'] = df['SM'].apply(star)
df['IR'] = df['IR'].apply(star)
```

The Hits column has some characters included in its values that needs to be fixed. The column also contain NaN values, we should change this values into 0.

```python
df.Hits = df.Hits.fillna(0)
```

```python
Hits = []

df['Hits'] = df['Hits'].astype(str)
```

```python
for i in df['Hits'] :
    if "K" in i:
        i = i.replace('K','')
        a = float(i) * 1000
        Hits.append(a)
    else:
        Hits.append(i)
```

```python
df['Hits'] = Hits
```

```python
df.info()
```

As we can see above, there are columns with incorrect data types. 

```python
df = df.astype({'Value in €':'int64'})
df = df.astype({'Wage in €' :'int64'})
df = df.astype({'Release Clause in €' :'int64'})
df = df.astype({'Year of Joining' :'int64'})
df = df.astype({'Day of Joining' :'int64'})
df = df.astype({'W/F' :'int64'})
df = df.astype({'SM' :'int64'})
df = df.astype({'IR' :'int64'})
```
